#Generar función para contar los digitos de la columna "NIT" a fin de verificar si hay registros con mayor a 9 digitos

```python
def contar_digitos(x):
    return len(str(x))
```

```python
longitud_nit = df["NIT"].apply(contar_digitos)==9
```

```python
print(longitud_nit)
```

#Verificar si todos los valores de la columna "NIT" cumplen con la condición

```python
if longitud_nit.all():
    print("Todos los valores de la columna 'NIT' cumplen con la condición.")
else:
    print("No todos los valores de la columna 'NIT' cumplen con la condición.")
```
